```python
class ElectronicDevice:

    def __init__(self, height, weight):
        
        self.height = height
        self.weight = weight
        self.alert()

    def alert(self):
        if self.weight > 100:
            print("!ALERT! - Devices is heavy! Do not lift without assistance!")
```

```python
class Computer(ElectronicDevice):

    def __init__(self, height, weight, mhz, gpu_mhz, ram, vram):
        super().__init__(height, weight)
        self.mhz = mhz
        self.gpu_mhz = gpu_mhz
        self.ram = ram
        self.vram = vram

    def change_component(self):
        
        component_choice = input("Please select a component: CPU, RAM or Graphics:") 
        
        if component_choice == 'CPU':
            self.change_cpu()
        elif component_choice == 'Graphics':
            self.change_graphics()
        elif component_choice == 'RAM':
            self.change_ram()
        else:
            print("Please choose from the listed selection.")        
            
        
    def change_graphics(self):
        new_gpu = int(input("How fast is your new graphics card?"))
        new_vram = int(input("How many gigs of VRAM does the new card have"))
        if new_vram > 8:
            print("Nice!")
        self.gpu = new_gpu
        self.vram = new_vram
        

    def change_cpu(self):
        new_mhz = int(input("How fast is your new CPU?"))
        self.mhz = new_mhz

    def change_ram(self):
        new_ram_amount = int(input("How many total GB now?"))
        new_ram_ghz = int(input("How fast is the RAM?"))
        self.ram = (new_ram_amount, new_ram_ghz)    
```
