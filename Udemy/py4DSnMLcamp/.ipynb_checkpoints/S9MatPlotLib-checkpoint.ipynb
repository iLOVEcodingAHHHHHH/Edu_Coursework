```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
x
```

```python
y=x**2
```

```python
plt.plot(x,y)
plt.title("Fun")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OOP
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,1,1])
axes2 = fig.add_axes([0.30,0.25,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Daddy')

axes2.plot(y,x)
axes2.plot('Baby')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=3)
plt.tight_layout()
#axes.plot(x,y) #subplots allow specific num columns
```

```python
axes
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.set_title('TitleString')
ax.set_xlabel('axislabelstring')
ax.plot(x,x**2,label='x squared')
ax.plot(y,x**3,label='x cubed')

ax.legend(loc=10)
```

```python
fig.savefig('my_picture.png')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='blue',lw=1,ls='-',marker='o')

ax.set_xlim([0,2])
ax.set_ylim([0,2])
```
