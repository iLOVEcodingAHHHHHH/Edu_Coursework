```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
```

```python
from collections import deque

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, value):
        new_node = Node(value)
        if self.root == None:
            self.root = new_node
            return self
        token = self.root
        while value != token.value:
            print(token.value)
            if value < token.value:
                if not token.left:
                    token.left = new_node
                    break
                token = token.left
            else:
                if not token.right:
                    token.right = new_node
                    break
                token = token.right
        if value == token.value:
            raise Exception('duplicates are not permitted')
        return self

    def search(self, value):
        if not self.head:
            raise Exception('Tree is empty.')
        token = self.head
        while token:
            if token.value == value:
                return True
            if value < token.value:
                token = token.left
            else:
                token = token.right
        return False

    def remove(self, value, start=None, parent=None):
        token = start or self.root
        while token and token.value != value:
            parent = token
            if value < token.value:
                token = parent.left
            else:
                token = parent.right
        if not token:
            raise Exception('item not in tree')
        if not token.right and not token.left:
            return self._remove_node_no_children(token, parent)
        if token.right and token.left:
            return self._remove_node_two_children(token)
        return self._remove_node_one_child(token, parent)

    def _remove_node_no_children(token, parent):
        if token is self.root:
            self.root = None
            return self
        if parent.left == token:
            parent.left = None
        else:
            parent.right = None
        return self

    def _remove_node_one_child(token, parent):
        if token == self.root:
            self.root = token.right or token.left
            return self
        child = token.left or token.right
        if parent.left == token:
            parent.left = child
        else:
            parent.right == child
        return self

    def _remove_node_two_children(token):
        successor = self._get_successor(token)
        token.value = successor.value
        return self.remove(successor.value, start=token.right, parent=token)
        
    @staticmethod
    def _get_successor(token):
        successor = token.right
        #why check for successor?
        while successor and successor.left:
            successor = successor.left
        return successor

    def breadth_first_traversal(self):
        if not self.root:
            raise Exception("Tree is empty")
        queue = deque()
        queue.append(self.root)
        visited = []
        while queue:
            visited_node = queue.popleft()
            visited.append(visited_node.value)
            if visited_node.left:
                queue.append(visited_node.left)
            if visited_node.right:
                queue.append(visited_node.right)
        return visited

    def dft_pre_order_iterative(self):
        if not self.root:
            raise Exception("Tree is empty")
        stack = [self.root]
        visited = []
        while stack:
            visited_node = stack.pop()
            visited.append(visited_node.value)
            if visited_node.left:
                stack.append(visited_node.left)
            if visited_node.right:
                stack.append(visited_node.right)
        return visited


    def dft_pre_order_recursive(self):
        if not self.root:
            raise Exception("Tree is empty")
        visited = []
        def _traverse(node):
            if node:
                visited.append(node.value)
                _traverse(node.left)
                _traverse(node.right)
            return
        _traverse(self.root)
        return visited

    def dft_in_order_iterative(self):
        if not self.root:
            raise Exception("Tree is empty")
        focus = self.root
        stack = []
        content = []
        while stack or focus:
            if focus:
                stack.append(focus)
                focus = focus.left
            else:
                visited_node = stack.pop()
                content.append(visited_node)
                if visited_node.right:
                    focus = visited_node.right
        return visited

    def dft_in_order_recursive(self):
        if not self.root:
            raise Exception("Tree is empty")
        visited = []
        focus = self.root
        def _traverse(node):
            if node:
                _traverse(focus.left)
                visited.append(node.value)
                _traverse(focus.right)
            return
        _traverse(self.root)
        return visited

    def dft_post_order_iterative(self):
        if not self.root:
            raise Exception("Tree is empty")
        previous = focus = self.root
        stack = []
        content = []
        while focus:
            while current.left:
                stack.append(focus)
                focus = focus.left
            while not focus.right or current.right == previous:
                visited.append(current.value)
                previous = current
                if not stack:
                    return visited
                current = stack.pop
            stack.append(current)
            current = current.right
        return content

        def dft_in_order_recursive(self):
        if not self.root:
            raise Exception("Tree is empty")
        visited = []
        focus = self.root
        def _traverse(focus):
            if focus:
                _traverse(focus.left)
                _traverse(focus.right)
                visited.append(focus.value)
            return
        _traverse(self.root)
        return visited
```
